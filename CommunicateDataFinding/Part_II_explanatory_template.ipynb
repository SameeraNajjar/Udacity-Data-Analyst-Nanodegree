## Investigation Overview


>This analysis's main objective is to look into the features of loans that the peer-to-peer lending platform Prosper issues. In particular, we want to know how different elements like the borrower's annual percentage rate (APR), Prosper rating, and debt-to-income ratio affect the loan's status and other consequences. The main conclusions drawn from this investigation are meant to offer useful data that lenders and borrowers on the platform can utilize.



## Dataset Overview and Executive Summary

>The dataset includes 81 variables, such as borrower details, loan terms, and loan performance, and includes data on 113,937 loans. Principal findings from the dataset consist of:

>Distribution of Borrower APR: There is a noticeable peak at 0.36, and borrower APRs are often grouped between 0.1 and 0.4.
Prosper Rating Distribution: Less loans are classified as AA or HR, with the majority of loans having a Prosper Rating in the middle categories (C, B, and A).
>Loan Performance: Higher rates of "Chargedoff" and "Defaulted" loan statuses are indicative of worse loan performance, which is correlated with higher debt-to-income ratios.





```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
loan_data = pd.read_csv('prosperLoanData.csv')
```

## (Distribution of Borrower APR)

> The distribution of Borrower APRs shows a right-skewed pattern, with most values between 0.1 and 0.4. The spike around 0.36 suggests that this rate is common among many borrowers.

```python
# Plotting the distribution of BorrowerAPR
plt.figure(figsize=(10, 6))
sns.histplot(loan_data['BorrowerAPR'].dropna(), kde=True, color='blue')
plt.title('Distribution of Borrower APR')
plt.xlabel('Borrower APR')
plt.ylabel('Frequency')
```

## (Distribution of Prosper Rating (Alpha))

> The most frequent Prosper Ratings are C, B, and A. This indicates that most borrowers fall within the mid-range of creditworthiness, with fewer borrowers at the extremes of AA (high creditworthiness) and HR (low creditworthiness).

```python
# Setting up the proper order for 'ProsperRating (Alpha)'
prosper_rating_order = ['AA', 'A', 'B', 'C', 'D', 'E', 'HR']
loan_data['ProsperRating (Alpha)'] = pd.Categorical(loan_data['ProsperRating (Alpha)'], categories=prosper_rating_order, ordered=True)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=loan_data, x='ProsperRating (Alpha)', palette='Blues_d', order=prosper_rating_order)
plt.title('Distribution of Prosper Rating (Alpha)')
plt.xlabel('Prosper Rating (Alpha)')
plt.ylabel('Count')
```

## (Debt to Income Ratio by Prosper Rating (Alpha) and Loan Status)

> This visualization highlights how the Debt-to-Income Ratio varies across different Prosper Ratings and loan statuses. It reveals that higher Debt-to-Income Ratios are more common among lower-rated loans (D, E, HR), especially those that have been charged off.

```python
plt.figure(figsize=(14, 8))
sns.violinplot(data=loan_data, x='ProsperRating (Alpha)', y='DebtToIncomeRatio', hue='LoanStatus', inner="quart", palette='Set3',)
plt.title('Debt to Income Ratio by Prosper Rating (Alpha) and Loan Status')
plt.ylabel('Debt to Income Ratio')
plt.xlabel('Prosper Rating (Alpha)')
plt.xticks(rotation=45)
plt.legend(loc='upper right')
```

### Conclusion

>Numerous significant insights regarding the relationship between loan results and borrower characteristics and loan terms have been uncovered by the analysis of the Prosper Loan data. Negative loan outcomes, such defaults or charge-offs, are more likely for borrowers with lower Prosper ratings and larger debt-to-income ratios. By using these information, the platform's lending decisions may be made with a better understanding of risk and an optimized loan performance.
